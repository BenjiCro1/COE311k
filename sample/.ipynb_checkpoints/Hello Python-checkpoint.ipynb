```python
import random
import math
def approxPi(samples):
    counter = 0
    for i in range(samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(math.sqrt(x*x+y*y) < 1):
            counter = counter + 1
    return (4 * (counter/samples))
```

```python
approxPi(1000000)
```
