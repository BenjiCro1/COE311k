```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def taylor_sin(x, n):
    out = 0.0
    for i in range(n):
        out += ((-1)**i) * (x**(2*i+1)) / math.factorial(2*i+1)
    return out
```

```python
steps = np.arange(0, 10.1, 0.1)
error1 = abs(np.sin(steps) - taylor_sin(steps,1))
error2 = abs(np.sin(steps) - taylor_sin(steps,2))
error3 = abs(np.sin(steps) - taylor_sin(steps,3))
error4 = abs(np.sin(steps) - taylor_sin(steps,4))
```

```python
plt.plot(steps, error1, label='One Term')
plt.plot(steps, error2, label='Two Terms')
plt.plot(steps, error3, label='Three Terms')
plt.plot(steps, error4, label='Four Terms')
plt.xlabel('X value inputed')
plt.ylabel('Error between calculated sin and true sin')
plt.legend()
plt.title('Taylor series sin Error')
plt.show()
```

Discussion: The truncation error decrease for the small valves of x inputed, as you add more terms. However, for large x values, greater terms actually result in more error and a less accurate result, so the change in error is not uniform.
