```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    start = time.perf_counter()
    n = len(b)
    x = x0.copy()
    errors = []
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i+1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        err = np.linalg.norm(x_new - x)
        errors.append(err)
        if err < tol:
            return x_new, k+1, time.perf_counter()-start, errors
        x = x_new
    return x, max_iterations, time.perf_counter()-start, errors
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    start = time.perf_counter()
    n = len(b)
    x = x0.copy()
    errors = []
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        err = np.linalg.norm(x_new - x)
        errors.append(err)
        if err < tol:
            return x_new, k+1, time.perf_counter()-start, errors
        x = x_new
    return x, max_iterations, time.perf_counter()-start, errors
```

```python
A = np.array([[3, -.1, -.2],
            [.1, 7, -.3],
            [.3, -.2, 10]], dtype=float)
b = np.array([7.85, -19.3, 71.4], dtype=float)
x0 = np.zeros_like(b)
tolerances = [1e-3, 1e-6, 1e-9]
out = {}

for tol in tolerances:
    gs = gauss_seidel(A, b, x0, tol, 100)
    jb = jacobi(A, b, x0, tol, 100)
    out[tol] = {"gs": gs, "jb": jb}

    print(f"Tolerance {tol}")
    print("GS iterations:", gs[1])
    print("Jacobi iterations:", jb[1])
    gs_errors = out[tol]["gs"][3]
    jb_errors = out[tol]["jb"][3]

    plt.figure()
    plt.plot(gs_errors, label="Gauss-Seidel")
    plt.plot(jb_errors, label="Jacobi")
    plt.xlabel("Iteration")
    plt.ylabel("Error")
    plt.title(f"Convergence with (tol={tol})")
    plt.legend()
    plt.show()
```

The greater precision required(lower tolerance) requires more iterations as while the changes between each iteration grow smaller and smaller, those changes are more significant with a lower tolerance so it needs more steps to slowly get there.
