```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
def approxPi(samples):
    counter = 0
    for i in range(samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(math.sqrt(x*x+y*y) < 1):
            counter = counter + 1
    return (4 * (counter/samples))
```

```python
samples = np.arange(10, 10001, 20)
error = np.zeros(len(samples))
for i in range(len(samples)):
    pi_est = approxPi(samples[i])
    error[i] = abs(pi_est - math.pi)


plt.plot(samples, error)
plt.xlabel('Number of Samples')
plt.ylabel('Error')
plt.title('Monte Carlo Approximation Error for Ï€')
plt.show()
```
