```python
import numpy as np
import time
```

```python
def gauss_seidel(A, b, x0, tol, max_iterations):
    start = time.perf_counter()
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            end = time.perf_counter() - start
            return x_new, k, end
        x = x_new
    end = time.perf_counter() - start
    return x, max_iterations, end
```

```python
def jacobi(A, b, x0, tol, max_iterations):
    start = time.perf_counter()
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            end = time.perf_counter() - start
            return x_new, k, end
        x = x_new
    end = time.perf_counter() - start
    return x, max_iterations, end
```

```python
A = np.array([
[0.582745, 0.48 , 0.10 , 0.  , 0.  ],
[0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
[0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
[0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
[0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)

b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations, time1 = gauss_seidel(A, b, x0, tol, max_iterations)
print(f"Solution for gauss: {solution}")
print(f"Iterations for gauss: {iterations}")
print(f"Time for gauss: {time1}")
solution2, iterations2, time2 = jacobi(A, b, x0, tol, max_iterations)
print(f"Solution for jacobi: {solution2}")
print(f"Iterations for jacobi: {iterations2}")
print(f"Time for jacobi: {time2}")
```

Jacobi does not solve, lack of iterations pervents
