```python
import numpy as np
```

```python
num = 1
summ = 0
for i in range(5000):
    if (i % 2 == 1):
        summ = summ - (1/num)
    else:
        summ = summ + (1/num)
    num = num + 2
    true = abs(np.pi-(4*summ))
    rel = abs(np.pi-(4*summ))/np.pi
    if (i % 100 == 0):
        print(true)
        print(rel)
```

As terms are added to the summation, the error decreases
