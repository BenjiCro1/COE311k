```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=0
b=2 * np.pi
N=50
x = np.linspace(a, b, N)

def f(x):
  return np.sin(x)

# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_dx_for = (f(x + h) - f(x))/h

# Backward Difference Approximation
df_dx_back = (f(x) - f(x - h))/h
```

```python
plt.plot(x, np.cos(x), label="cos(x)")
plt.plot(x, df_dx, label="Central Difference")
plt.plot(x, df_dx_for, label="Forward Difference")
plt.plot(x, df_dx_back, label="Backward Difference")
plt.xlabel("x")
plt.ylabel("Derivative")
plt.title("Derivative Approximations vs True Derivative")
plt.legend()
plt.grid()
plt.show()
```
