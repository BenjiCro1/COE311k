```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
Y_guess = (m * X) + c
plt.scatter(X, Y, label="Data Points")
plt.plot(X, Y_guess, label="Best Fit Line", linewidth=2)

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Least Squares Fit Part One")
plt.legend()
plt.show()
```

```python
#Part 2
X = np.random.rand(10) * 10
X = np.sort(X)
Y = np.random.rand(10) * 10
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
Y_guess = (m * X) + c
plt.scatter(X, Y, label="Data Points")
plt.plot(X, Y_guess, label="Best Fit Line", linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Least Squares Fit Part Two")
plt.legend()
plt.show()  #Model no longer represents data well
```

```python
#Part 3
X = np.arange(0, 1.01, 0.01)
m = np.random.rand()
b = np.random.rand()
Y = (m * X) + b
Z = (np.random.rand(101) * 2) - 1
Y_perm = (m * X) + b + Z
A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y_perm, rcond=None)[0]
print(f"Slope: {m}, Intercept: {c}")
```

```python
Y_guess = (m * X) + c
res = Y_perm - Y_guess
max_res = np.argmax(np.abs(res))
min_res = np.argmin(np.abs(res))
print("Largest residual error at X = .", max_res, " with error of ", abs(res[max_res]))
print("Smallest residual error at X = .", min_res, " with error of ", abs(res[min_res]))
plt.scatter(X, Y_perm, label="Data Points")
plt.plot(X, Y, label="Actual Y")
plt.plot(X, Y_guess, label="Best Fit Line", linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Least Squares Fit Part Three")
plt.legend()
plt.show()
```
