```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt
def approxPi(samples):
    counter = 0
    for i in range(samples):
        x = random.uniform(-1,1)
        y = random.uniform(-1,1)
        if(math.sqrt(x*x+y*y) < 1):
            counter = counter + 1
    return (4 * (counter/samples))
```

```python
for i in range(100):
    cur = approxPi(1000)
    true = abs(np.pi-cur)
    rel = abs(np.pi-cur)/np.pi
    plt.scatter(i, true, 20, 'red')
    plt.scatter(i, rel, 20, 'blue')
```
