```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.subplot(2, 2, 1)
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 2, 2)
plt.scatter(x, np.cos(x), 20)
x = np.random.rand(1000)
plt.subplot(2, 2, 3)
plt.hist(x, 50)
plt.subplot(2, 2, 4)
z1 = np.random.randint(0, 1001, size=(2001, 2001))
plt.imshow(z1, cmap='hot')
plt.show()
```
