```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def naive_matrix_multiplication(A, B):
  n = len(A)
  C = [[0] * n for _ in range(n)]
  for i in range(n):
    for j in range(n):
      for k in range(n):
        C[i][j] += A[i][k] * B[k][j]
  return C

custom_time = []
numpy_time = []
sizes = []
for i in range(8):
  size = 3 * (2 ** i)
  print(f"{size}x{size}")

  A = np.random.rand(size, size)
  B = np.random.rand(size, size)

  start = time.time()
  naive_matrix_multiplication(A, B)
  end = time.time()
  custom_time.append(end - start)

  start = time.time()
  np.dot(A, B)
  end = time.time()
  numpy_time.append(end - start)

  sizes.append(size)
```

```python
plt.plot(sizes, custom_time, label='Custom')
plt.plot(sizes, numpy_time, label='NumPy')
plt.xlabel('Matrix size(value by value)')
plt.ylabel('Runtime')
plt.title('Runtime Comparison')
plt.legend()
plt.grid()
plt.show()
```
