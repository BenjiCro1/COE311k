```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parameters
k = 1.0
h =  [.01, .02, .03, .05, .1, .2, .3, .5]
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

plt.figure(figsize=(10, 5))

for step in h:
    x_values = np.arange(x_start, x_end + step, step)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0

    # Euler's Method loop

    for i in range(1, len(x_values)):
        # Current values
        x_n = x_values[i - 1]
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        # Update equations
        y_values[i] = y_n + step * v_n
        v_values[i] = v_n + step * (-k * y_n)

    plt.plot(x_values, y_values, label=f'h = {step}')

plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('Solution of d²y/dx² = -ky using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```

Looks like it becomes unstable at h=.1, with eveything past that looking exponentially worse as the function progresses. When k is at 10 the function occilates far faster so the exponential changes get far worse and the function gets tens of thousands off where it should be
