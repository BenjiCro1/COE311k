```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parameters
k = 1.0
h =  [.1, .01, .001, .00001]
t_start = 0
t_end = 30
y_0 = 1.0

plt.figure(figsize=(10, 5))

for step in h:
  t_values = np.arange(t_start, t_end + step, step)
  y_values = np.zeros(len(t_values))
  y_values[0] = y_0

    # Euler's Method loop
  for i in range(1, len(t_values)):
      # Current values
      t_n = t_values[i - 1]
      y_n = y_values[i - 1]

      # Update equations
      y_values[i] = y_n + step * (-1000 * y_n + 3000 - 2000 * np.exp(-1 * t_n))
  plt.plot(t_values, y_values, label=f'h = {step}')

plt.plot(t_values, 3 - .998 * np.exp(-1000*t_values) - 2.002 * np.exp(-t_values), label=f"analytical")

plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution of dy/dt = -1000y+3000-2000*e^(-t) using Euler\'s Method')
plt.legend()
plt.grid(True)
plt.show()
```
